```python
from google.colab import files
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/DS/Final/Data/recipes.csv")
```

擷取有用的資訊

```python
df = df[['recipe_name', 'nutrition', 'img_src', 'url']]
df = df.rename(columns={'recipe_name': 'name'})

# 查看處理後的資料
df.head()
```

截取食譜內的營養標示

```python
import re

# 定義一個函數來解析 nutrition 欄位並提取所需的數據
def extract_value(pattern, text):
    """根據提供的 pattern 從字串中提取數值"""
    match = re.search(pattern, text)
    if match:
        return float(match.group(1))
    return 0.0

# 定義欄位擷取的正則表達式
patterns = {
    'Calories_Kcal': lambda x: (
        extract_value(r"Total Fat (\d+)g", x) * 9 +
        extract_value(r"Total Carbohydrate (\d+)g", x) * 4 +
        extract_value(r"Protein (\d+)g", x) * 4
    ),
    'Carbs_g': r"Total Carbohydrate (\d+)g",
    'Cholesterol_mg': r"Cholesterol (\d+)mg",
    'Fats_g': r"Total Fat (\d+)g",
    'Fiber_g': r"Dietary Fiber (\d+)g",
    'Protein_g': r"Protein (\d+)g",
    'Sugars_g': r"Total Sugars (\d+)g"
}

# 新增欄位
for col, pattern in patterns.items():
    if callable(pattern):  # 如果 pattern 是函數
        df[col] = df['nutrition'].apply(pattern)
    else:  # 如果 pattern 是正則表達式
        df[col] = df['nutrition'].apply(lambda x: extract_value(pattern, x))

# 刪除原始 nutrition 欄位
df = df.drop(columns=['nutrition'])

# 調整欄位順序
columns_order = ['name', 'Calories_Kcal', 'Carbs_g', 'Cholesterol_mg', 'Fats_g', 'Fiber_g', 'Protein_g', 'Sugars_g', 'img_src', 'url']
df = df[columns_order]

df.head()

# 如果需要保存處理後的資料為新的 CSV
# df.to_csv('processed_recipes.csv', index=False)
```
